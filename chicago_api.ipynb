```python
painting_ids = []

for artwork in range(1,200000,1):
    try:
        f = open(f'/artipy/artic-api-data/json/artworks/{artwork}.json')
        data = json.load(f)
        if  "Pop Art" in data['style_title']:
            painting_ids.append(artwork)
            f.close
        else: f.close()
    except: artwork =  artwork +1
```
